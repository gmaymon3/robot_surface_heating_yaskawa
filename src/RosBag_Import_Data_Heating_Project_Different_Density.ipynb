```python
# Import math and csv
import math
import csv
# Enter size of heated node grid
# For Example for a heated node grid of 5x5 (total temperature array of 7x7) enter the following
# heated_node_size = 5

heated_node_size = 5


###################   PROCESS END EFFECTOR (EE_NODE) DATA  ############################
# End effector data
ee_data = b.message_by_topic('/EE_node')
print("File saved: {}".format(ee_data))
ee_data_csv = pd.read_csv(ee_data)
start_time = ee_data_csv.at[0,'Time']
ee_data_list = ee_data_csv.values.tolist()

# Send each data point to corresponding list
ee_time_list = [data_point[0] for data_point in ee_data_list]
x_list = [data_point[3] for data_point in ee_data_list]
y_list = [data_point[4] for data_point in ee_data_list]
print(ee_time_list)


####################   PROCESS TEMPERATURE ARRAY (TEMP_ARRAY) DATA  ####################
# Temperature Data
temp_data = b.message_by_topic('/Temp_Array')
print("File saved: {}".format(temp_data))
temp_data_csv = pd.read_csv(temp_data)
end_time_ind = len(temp_data_csv)
x_value_length = int(math.sqrt(np.shape(temp_data_csv)[1]-3)) # first 3 elements are not part of temp grid
y_value_length = x_value_length # Assume the grid is square

# Get start and end time
end_time = temp_data_csv.at[end_time_ind-1,'Time']
start_time = temp_data_csv.at[0,'Time']

# Process temperature data and create a list of each data point
temp_data_list = temp_data_csv.values.tolist()
temp_arrays = []
for idx_temp, data_p in enumerate(temp_data_list):
  temp_array = np.reshape(temp_data_list[idx_temp][3:],(x_value_length,y_value_length))
  temp_arrays.append(temp_array)

temp_time_list = [data_point[0] for data_point in temp_data_list]
print(temp_time_list)

# Show an example plot of the temperature data for a sanity check
cmap = plt.cm.viridis
plt.imshow(temp_arrays[244], cmap=cmap) # 244 is arbitrary
plt.colorbar()
plt.show()



################## ADJUST START TIME SO WE HAVE EE DATA #############################
# Adjust this so difference is positive and the lists are shortened such that temperature list always has EE position information
print("Temperature Start Time")
print(temp_time_list[0])
print("End Effector Start Time")
print(ee_time_list[0])
print("Differece: ")
print(temp_time_list[30]-ee_time_list[0])

# Shorten lists
start_val = 30
temp_arrays = temp_arrays[start_val:]
temp_time_list = temp_time_list[start_val:]
ee_time_numpy = np.array(ee_time_list)


######################   FLIP Y SO 1 IS AT TOP   ####################################
# Flip the y such that 1 becomes 5 for example. This is becuase we want 5 to be the bottom instead of the top like before.
reverse_list = np.linspace(heated_node_size, 1, heated_node_size).tolist()

y_list_flip = []
# Create flipped y list
for y in y_list:
  if y == 0:
    y_list_flip.append(0)
  else:
    y_list_flip.append(int(reverse_list[y-1]))

x_list_flip = []
# Create flipped x list
for x in x_list:
  if x == 0:
    x_list_flip.append(0)
  else:
    x_list_flip.append(int(reverse_list[x-1]))


# Note: THIS ASSUMES A 25 x 25 grid for temp and 5x5 for heated node

dif_den_h_node_list = [[53,58,63,68,73],[178,183,188,193,198],[303,308,313,318,323],[428,433,438,443,448],[553,558,563,568,573]]

# Reformat heated pos nodes
heated_node_list = []
for i,x_val in enumerate(x_list):
  if x_list[i] == 0:
    heated_node = -1
  else:
    print(x_list[i])
    print(y_list[i])
    heated_node = dif_den_h_node_list[x_list[i]-1][y_list[i]-1]
    print(heated_node)
  heated_node_list.append(heated_node)

print(heated_node_list)
print(x_list)
print(y_list_flip)

##############    REFORMAT TEMPERATURES INTO NODE ORDER AS SHOWN    #######################
# Reformat the temperature node order
# Example below
# 10 11 12 13 14
# 15  1  2  3 16
# 17  4  5  6 18
# 19  7  8  9 20
# 21 22 23 24 25

temp_lists = np.empty((0, x_value_length*y_value_length)) # Create empty temp lists

for index,temp_array_ex in enumerate(temp_arrays):

  temp_list = []

  #print(temp_array_ex)
  # Get Heated Node Temps
  heated_node_temps = temp_array_ex[1:x_value_length-1,1:y_value_length-1]

  # Get Top Edge Temps
  top_edge_temps = temp_array_ex[0,:]

  # Get Side Edge Temps
  left_edge_temps = temp_array_ex[1:y_value_length-1,0]
  right_edge_temps = temp_array_ex[1:y_value_length-1,x_value_length-1]

  edge_list_temps = []
  # Assign in alternating order to list
  for value1, value2 in zip(left_edge_temps, right_edge_temps):
      edge_list_temps.append(value1)
      edge_list_temps.append(value2)

  # Get Bottom Edge Temps
  bottom_edge_temps = temp_array_ex[y_value_length-1,:]

  temp_list.extend(heated_node_temps.flatten())
  temp_list.extend(top_edge_temps.flatten())
  temp_list.extend(np.array(edge_list_temps).flatten())
  temp_list.extend(bottom_edge_temps.flatten())
  temp_lists = np.vstack((temp_lists, temp_list)) # Stack this new list on the old lists

  if index % 1000 == 0:
    print(index) # Show progress

print(np.shape(temp_lists))

################## CREATE ARRAY FOR FULL DATA SET  #############################
# data_arrays contians a list of lists with time, temperature array, and heated node
data_arrays = np.empty((0, 2+x_value_length*y_value_length))

for idx, time in enumerate(temp_time_list):
  # Create new array
  ee_index = np.sum(ee_time_numpy < temp_time_list[idx])-1
  temp_array_at_t = temp_lists[idx]
  node_array_at_t = np.array([heated_node_list[ee_index]])
  data_array = np.concatenate((np.array(time).reshape((1,)), temp_array_at_t,node_array_at_t)) # time, temperature array, and heated node
  data_arrays = np.vstack((data_arrays,data_array))

  if idx % 100 == 0:
    print(node_array_at_t)
    cmap = plt.cm.viridis
    plt.imshow(temp_arrays[idx], cmap=cmap)
    plt.colorbar()
    plt.show()

min_time = min(data_arrays[:,0])
data_arrays[:,0] = data_arrays[:,0]-min_time
#print(data_arrays[:,0])

################## CREATE DATA NAMES FOR EACH VARIABLE  #############################
# Using a loop to create a list of strings
node_list = []
for i in range(x_value_length*y_value_length):
    node_list.append(f"node{i}")


# Add header list to data arrays
print(node_list)
header_list = []
header_list.append('time')
header_list.extend(node_list)
header_list.append('heating_node')
print(np.shape(header_list))
data_arrays = np.vstack([header_list, data_arrays])

print(np.shape(node_array_at_t))
print(np.shape(temp_array_at_t))
print(np.shape(data_array))
print(np.shape(data_arrays))

#### ADJUST THIS AS NEEDED ####
# Specify the file path for the CSV file
csv_file_path = "heating_data_25x25temp_5x5heated.csv"

# Save the array to a CSV file
# Write the NumPy array with headers to a CSV file
with open(csv_file_path, mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerows(data_arrays)


```
